 # Model Creation - imp .step 

```python
df.head()
```

```python
# step1: Indepedent and dependent variable creation 
```

```python
x= df[['TV','Radio','Newspaper']] # x contain all inputs 
y=df['Sales'] # y contain output i.e target
```

```python
x # Indepedent features varaibles or features
```

```python
y # series  
# target 
```

```python
type(y)
```

```python
# 1. training and testing data
```

```python
from sklearn.model_selection import train_test_split # split the data
# training data and test data 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train # training variables 
```

```python
y_train
```

```python
x_test # testing varabiles take from dataset 
```

```python
y_test # target test
```

```python
# step3 : creating the model 
from sklearn.linear_model import LinearRegression 
# actually use algorithm
```

```python
# model for linear regression 
LR= LinearRegression()#object creation
# apply lr algo
```

```python
LR
```

```python
LR.fit(x_train,y_train) # trainig the model
```

```python
                                                                                                                                                                                                                                                                                        y_pred=LR.predict(x_test) # getting predictions  # on testing data 
```

```python
y_test # actual 
```

```python
y_pred
```

```python
# check accuracy : 
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred) # use for classification 
```

```python
# model evaluation   # check performance mean .mse, mae,r2 score check the performance of our model
```

```python
# in linear regression evalueates ,mean square error ,mean absolute error, r2 score
```

- R2 score -explains how much variablity in target can be explain in model
